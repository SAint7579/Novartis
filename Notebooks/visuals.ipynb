```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Remove the Sample_id column for scaling (assuming it's the first column)
data_for_scaling = dataframe.iloc[:, 1:]  # Skip the first column (Sample_id)

# Min-max scale the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data_for_scaling)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
tsne_results = tsne.fit_transform(scaled_data)

# Create the 2D t-SNE plot
plt.figure(figsize=(10, 8))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1], alpha=0.6)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('2D t-SNE Plot of Min-Max Scaled Data')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Load R data file (.RData) and convert to pandas DataFrames
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter
import pandas as pd

# Use the new context-based conversion approach instead of deprecated activate()
with localconverter(robjects.default_converter + pandas2ri.converter):
    # Load the RData file
    robjects.r['load']("C:/Users/vishw/OneDrive/Desktop/Projects/Novartis/DRUG-seq/data/Novartis_drugseq_U2OS_MoABox/drugseq_ensembl_v98_annotation_and_entrez_mapping.RData")
    
    # List all objects in the R environment to see what was loaded
    r_objects = robjects.r['ls']()
    print("Objects loaded from RData file:")
    for obj in r_objects:
        print(f"- {obj}")
    
    # Convert all R objects to pandas DataFrames
    dataframes = {}
    for obj_name in r_objects:
        try:
            # Get the R object
            r_obj = robjects.r[obj_name]
            
            # Check if it's already a pandas DataFrame (converted automatically)
            if hasattr(r_obj, 'shape') and hasattr(r_obj, 'columns'):
                # It's already a pandas DataFrame
                dataframes[obj_name] = r_obj
                print(f"Successfully accessed {obj_name} as DataFrame with shape: {r_obj.shape}")
            else:
                # Try to convert using robjects conversion
                df = robjects.conversion.rpy2py(r_obj)
                dataframes[obj_name] = df
                print(f"Successfully converted {obj_name} to DataFrame with shape: {df.shape}")
        except Exception as e:
            print(f"Could not convert {obj_name} to DataFrame: {e}")
            # Try alternative conversion method
            try:
                # Direct access without conversion
                dataframes[obj_name] = r_obj
                print(f"Stored {obj_name} as R object (use robjects methods to access)")
            except Exception as e2:
                print(f"Could not store {obj_name}: {e2}")
    
    # Access individual DataFrames like: dataframes['object_name']
    # Example: annotation_df = dataframes['drugseq_ensg_v98']
```
